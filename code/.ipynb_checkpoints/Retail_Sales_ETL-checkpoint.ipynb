```python
import pandas as pd
file_path =r"C:\Users\Jina\Documents\Retail-Sales\Data\Original_data.xlsx"
excel = pd.ExcelFile(file_path)


retail1 = pd.read_excel(file_path,sheet_name="retail_data1")

retail2 = pd.read_excel(file_path,sheet_name="retail_data2")

products = pd.read_excel(file_path,sheet_name="product_details")
```

```python
retail = pd.concat([retail1, retail2],ignore_index=True)

print(retail.shape)
```

```python
retail['product_name'] = ( retail['product_name'].str.strip().str.title())

products['product_name'] = ( products['product_name'] .str.strip() .str.title())
```

```python
category_map = {
    'CLOTH':'Clothing',
    'clothing':'Clothing',

    'ELEC':'Electronics',
    'electronics':'Electronics',

    'FURN':'Furniture',
    'furniture':'Furniture',

    'HOME':'Home Appliances',
    'home appliances':'Home Appliances'
}
```

```python
retail['category'] = retail['category'].replace(category_map)
```

```python
product_price_map = (
    products
    .set_index('product_id')['price']
)
```

```python
retail['price'] = retail['price'].fillna(retail['product_id'].map(product_price_map))
```

```python
before = len(retail)

retail = retail[retail['quantity'] > 0]

after = len(retail)

print("Records Removed:",before-after)
```

```python
duplicate_txns = retail[ retail['transaction_id'].duplicated(keep=False)].sort_values('transaction_id')
duplicate_txns.head(20)
```

```python
before = len(retail)

retail = retail.drop_duplicates(subset=['transaction_id'],keep='first')
after = len(retail)
print("Removed:", before-after)
```

```python
retail['transaction_id'].duplicated().sum()
```

```python
retail['transaction_date'] = pd.to_datetime( retail['transaction_date'],errors='coerce')
```

```python
retail['transaction_date'].isnull().sum()
```

```python
def mask_email(email):
    if pd.isna(email):
        return email

    name, domain = email.split('@')
    return name[0] + "***@" + domain

retail['email'] = retail['email'].apply(mask_email)
```

```python
def mask_phone(phone):
    if pd.isna(phone):
        return phone

    phone = str(phone)
    return "******" + phone[-4:]

retail['phone'] = retail['phone'].apply(mask_phone)
```

```python
retail['revenue'] = (retail['price'] * retail['quantity']* (1 - retail['discount']))
```

### Cleaning Step 1: Category Standardization

Category values were standardized using a mapping dictionary to eliminate inconsistent naming conventions and abbreviations.

Result:
- Reduced multiple category variants into four standardized categories:
  - Clothing
  - Electronics
  - Furniture
  - Home Appliances

### Cleaning Step 2: Missing Price Imputation

A total of 809 missing price values were identified across the retail transaction datasets.

Resolution:
The product_details table was used as the authoritative source of product pricing. Missing values were populated using product_id based lookup.

Result:
- Missing price count reduced from 809 to 0.

### Cleaning Step 3: Invalid Quantity Removal

A total of 93 records contained invalid quantity values (0, -1, -2).

Business Rule:
Quantity must be greater than zero for a valid sales transaction.

Result:
- 93 invalid records removed.

### Cleaning Step : Duplicate Transaction Removal

Observation:
487 transaction identifiers appeared more than once in the dataset.

Investigation:
Records sharing the same transaction_id were analyzed and found to contain identical values across all business attributes including:

Conclusion:
These records represented duplicated transactions rather than valid multi-product orders.

Action Taken:
Duplicate transaction records were removed by retaining the first occurrence of each transaction_id.

Result:
487 duplicate transaction records were removed.

### Revenue Calculation
The discount field was observed to be stored as a decimal fraction (e.g., 0.05 representing 5%).

Therefore, transaction revenue was calculated as:

Revenue = Price × Quantity × (1 − Discount)

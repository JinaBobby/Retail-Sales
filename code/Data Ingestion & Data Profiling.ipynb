### Data Quality Issue 2: Invalid Quantities

Observation:
Non-positive quantity values (0, -1, -2) were identified in the transaction datasets.

Business Impact:
These records would distort revenue calculations, product performance metrics, and sales reporting.

Resolution:
Records with quantity <= 0 will be removed from the dataset during the cleaning phase.

### Data Quality Issue 3: Inconsistent Category Values

Observation:
The category column contains multiple representations of the same category such as:
- CLOTH, Clothing, clothing
- ELEC, Electronics, electronics
- FURN, Furniture, furniture
- HOME, Home Appliances, home appliances

Business Impact:
Revenue and sales metrics would be fragmented across multiple category labels, leading to inaccurate reporting.

Resolution:
Standardize category values using a mapping dictionary and convert them to a common format.

### Data Quality Issue 4: Inconsistent Product Names

Observation:
The same products appear in multiple formats.

Examples:
- LAPTOP, Laptop, laptop
- PHONE, Phone, phone
- DINING TABLE, Dining table, dining table

Business Impact:
Product-level sales and revenue analysis would be fragmented across multiple labels.

Resolution:
Standardize product names using the product dimension table and convert values to a consistent format.
